## Data Preparation

```python
#Create dummy variables for categorical variables
season = pd.get_dummies(bike['season'],drop_first=True)
weather= pd.get_dummies(bike['weathersit'],drop_first=True)
month= pd.get_dummies(bike['month'],drop_first=True)
weekday= pd.get_dummies(bike['weekday'],drop_first=True)

bike= pd.concat([bike,season,weather,month,weekday],axis=1)
bike.head()
```

```python
#Drop the categorical variables as they have been dummy encoded
bike.drop(['season','weathersit','weekday','month'],axis=1,inplace=True)

bike.head()
```

### Split the data to training and testing set

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
bike_train, bike_test = train_test_split(bike, train_size = 0.7, test_size = 0.3, random_state = 100)
```

###  Scaling the features

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

#Apply scaling on numeric variables leaving the dummy variables
num_vars=['temp','humidity','windspeed','count']

#Fit on data
bike_train[num_vars] = scaler.fit_transform(bike_train[num_vars])
bike_train.head()
```

### Do a heatmap and check for correlation


```python

plt.figure(figsize=(25, 20))
sns.heatmap(bike_train.corr(),annot = True)
plt.show()
```

## Building the Linear Model

```python
#Divide in to X and y
y_train = bike_train.pop('count')
X_train = bike_train
```

### Model1 Use Recursive Feature Elimination (RFE) for feature selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm,n_features_to_select=10) 

rfe = rfe.fit(X_train, y_train)


list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#pick the variables selected by RFE
rfe_cols = X_train.columns[rfe.support_]
rfe_cols
```

```python
X_train_rfe = X_train[rfe_cols]
```

```python
import statsmodels.api as sm  
X_train_lm1 = sm.add_constant(X_train_rfe)
lm1 = sm.OLS(y_train,X_train_lm1).fit()  
lm1.summary()
```

### Check Variance Inflation Factor (VIF)

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculateVif(X):
      vif = pd.DataFrame()
      vif['Features'] = X.columns
      vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
      vif['VIF'] = round(vif['VIF'], 2)
      vif = vif.sort_values(by = "VIF", ascending = False)
      return vif           

vif = calculateVif(X_train_rfe)

vif
```

## Model2 Drop feature humidity with high VIF

```python
#Drop humidity with high VIF
X_train2 = X_train_rfe.drop(["humidity"], axis = 1)
X_train_lm2 = sm.add_constant(X_train2)
lm2 = sm.OLS(y_train,X_train_lm2).fit()
print(lm2.summary())
```

```python
#Calculate VIF

vif = calculateVif(X_train2)

vif
```

### Model Fit
Now the all the VIF values are < 5 and p values are < 0.05 and the model is good to go

```python
#plt.scatter(X_train, y_train)
#plt.show
y_train.size
```

## Residual Analysis

```python
y_train_pred = lm2.predict(X_train_lm2)
fig = plt.figure()
sns.distplot((y_train - y_train_pred), bins = 20)
fig.suptitle('Residual Analysis', fontsize = 15)                 
plt.xlabel('Error', fontsize = 10)     

plt.show()
```

```python
plt.scatter(y_train_pred, y_train - y_train_pred)
plt.show()
```

### Making Prediction

```python
#Do the same scaling on the test set

#Fit on data
bike_test[num_vars] = scaler.fit_transform(bike_test[num_vars])
bike_test.head()
```

```python
#Divide in to X and y

y_test = bike_test.pop('count')
X_test = bike_test


X_test_lm = X_test[X_train2.columns]

# Adding a constant variable 
X_test_lm = sm.add_constant(X_test_lm)

y_test_pred = lm2.predict(X_test_lm)

fig = plt.figure()
plt.scatter(y_test,y_test_pred)
fig.suptitle('Actual vs Predicted', fontsize=15)            
plt.xlabel('Actual', fontsize=10)                         
plt.ylabel('Predicted', fontsize=10)   

plt.show()
```

```python
#Calculate r2
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, y_test_pred)

print('r2 = ', r2_score)
#Calculate adjusted r2 score
#adjusted_r2=1-(1-R2)*(n-1)/(n-p-1)

#n =sample size , p = number of independent variables

n = X_test.size
p = X_test_lm.columns.size-1 #Substract the constant)
print("n = ",n)
print("p = ",p)
adj_r2=1-(1-r2_score)*(n-1)/(n-p-1)
print("adjusted r2 = ",adj_r2)
```

The difference between the r2 score of test and train set is around 0.04 and we can say the model has generalized well
